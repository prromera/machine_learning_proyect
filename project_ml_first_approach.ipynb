```python
home_cols_new = [col + "_rolling" for col in home_cols]
```

```python
away_cols_new = [col + "_rolling" for col in away_cols]
```

```python
for season in df_new["season"].unique():
    for team in df_new[df_new["season"] == season].home_team.unique():
        for round in range(1,39):
            if season == 2024 and round >= 30:
                break
            if round == 1:
                query_result_home = df_new[(df_new["round"] == 1) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                query_result_away = df_new[(df_new["round"] == 1) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                if query_result_home.empty:
                    index = df_new[(df_new["round"] == 1) & (df_new["away_team"] == team) & (df_new["season"] == season)].index[0]
                    round_one_stats = [df_new.loc[index, source_col] for source_col in away_cols]
                    
                else:
                    index = df_new[(df_new["round"] == 1) & (df_new["home_team"] == team) & (df_new["season"] == season)].index[0]
                    round_one_stats = [df_new.loc[index, source_col] for source_col in home_cols]
                continue
                
            if round == 2:
                query_result_home = df_new[(df_new["round"] == 2) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                query_result_away = df_new[(df_new["round"] == 2) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                
                if query_result_home.empty:
                    index= df_new[(df_new["round"] == 2) & (df_new["away_team"] == team) & (df_new["season"] == season)].index[0]
                    round_two_stats = [df_new.loc[index, source_col] for source_col in away_cols]
                    df_new.loc[index, away_cols_new] = round_one_stats                   

                else:
                    index= df_new[(df_new["round"] == 2) & (df_new["home_team"] == team) & (df_new["season"] == season)].index[0]
                    round_two_stats = [df_new.loc[index, source_col] for source_col in home_cols]
                    df_new.loc[index, home_cols_new] = round_one_stats 
                continue
            
            if round == 3:
                query_result_home = df_new[(df_new["round"] == 3) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                query_result_away = df_new[(df_new["round"] == 3) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                
                if query_result_home.empty:
                    index= df_new[(df_new["round"] == 3) & (df_new["away_team"] == team) & (df_new["season"] == season)].index[0]
                    round_three_rolling = [ (x + y)/2 for x, y in zip(round_one_stats, round_two_stats)]
                    df_new.loc[index, away_cols_new] = round_three_rolling                    

                else:
                    index= df_new[(df_new["round"] == 3) & (df_new["home_team"] == team) & (df_new["season"] == season)].index[0]
                    round_three_rolling = [ (x + y)/2 for x, y in zip(round_one_stats, round_two_stats)]
                    df_new.loc[index, home_cols_new] = round_three_rolling
                continue
            else:
                query_result_home = df_new[(df_new["round"] == round) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                query_result_away = df_new[(df_new["round"] == round) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                
                if query_result_home.empty:

                    index= df_new[(df_new["round"] == round) & (df_new["away_team"] == team) & (df_new["season"] == season)].index[0]
                    query_one_home = df_new[(df_new["round"] == round-3) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                    query_one_away = df_new[(df_new["round"] == round-3) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                    query_two_home = df_new[(df_new["round"] == round-2) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                    query_two_away = df_new[(df_new["round"] == round-2) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                    query_three_home = df_new[(df_new["round"] == round-1) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                    query_three_away = df_new[(df_new["round"] == round-1) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                                    
                    if query_one_home.empty:

                        first_index = df_new[(df_new["round"] == round-3) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols].index[0]
                        round_x_stats= [df_new.loc[first_index, source_col] for source_col in away_cols]
                    else:
                        first_index = df_new[(df_new["round"] == round-3) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols].index[0]
                        round_x_stats= [df_new.loc[first_index, source_col] for source_col in home_cols]
                                     
                    if query_two_home.empty:
                        second_index = df_new[(df_new["round"] == round-2) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols].index[0]
                        round_y_stats= [df_new.loc[second_index, source_col] for source_col in away_cols]
                    else:
                        second_index = df_new[(df_new["round"] == round-2) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols].index[0]
                        round_y_stats= [df_new.loc[second_index, source_col] for source_col in home_cols] 
                                     
                    if query_three_home.empty:
                        third_index = df_new[(df_new["round"] == round-1) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols].index[0]
                        round_z_stats= [df_new.loc[third_index, source_col] for source_col in away_cols]
                    else:
                        third_index = df_new[(df_new["round"] == round-1) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols].index[0]
                        round_z_stats= [df_new.loc[third_index, source_col] for source_col in home_cols]                                      
                        
                    
                    rolling_avg = [ (x + y + z)/3 for x, y, z in zip(round_x_stats, round_y_stats, round_z_stats)]
                    df_new.loc[index, away_cols_new] = rolling_avg                

                else:
                    index= df_new[(df_new["round"] == round) & (df_new["home_team"] == team) & (df_new["season"] == season)].index[0]
                    query_one_home = df_new[(df_new["round"] == round-3) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                    query_one_away = df_new[(df_new["round"] == round-3) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                    query_two_home = df_new[(df_new["round"] == round-2) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                    query_two_away = df_new[(df_new["round"] == round-2) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                    query_three_home = df_new[(df_new["round"] == round-1) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols]
                    query_three_away = df_new[(df_new["round"] == round-1) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols]
                                    
                    if query_one_home.empty:
                        first_index = df_new[(df_new["round"] == round-3) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols].index[0]
                        round_x_stats= [df_new.loc[first_index, source_col] for source_col in away_cols]
                    else:
                        first_index = df_new[(df_new["round"] == round-3) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols].index[0]
                        round_x_stats= [df_new.loc[first_index, source_col] for source_col in home_cols]
                                     
                    if query_two_home.empty:
                        second_index = df_new[(df_new["round"] == round-2) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols].index[0]
                        round_y_stats= [df_new.loc[second_index, source_col] for source_col in away_cols]
                    else:
                        second_index = df_new[(df_new["round"] == round-2) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols].index[0]
                        round_y_stats= [df_new.loc[second_index, source_col] for source_col in home_cols] 
                                     
                    if query_three_home.empty:
                        third_index = df_new[(df_new["round"] == round-1) & (df_new["away_team"] == team) & (df_new["season"] == season)][away_cols].index[0]
                        round_z_stats= [df_new.loc[third_index, source_col] for source_col in away_cols]
                    else:
                        third_index = df_new[(df_new["round"] == round-1) & (df_new["home_team"] == team) & (df_new["season"] == season)][home_cols].index[0]
                        round_z_stats= [df_new.loc[third_index, source_col] for source_col in home_cols]                                      
                        
                    
                    rolling_avg = [ (x + y + z)/3 for x, y, z in zip(round_x_stats, round_y_stats, round_z_stats)]
                    df_new.loc[index, home_cols_new] = rolling_avg      
```
